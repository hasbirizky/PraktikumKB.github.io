# Menampilkan Informasi Jumlah Record, Jumlah Attribute, Nama Attribute, Jumlah Attribute Dengan Tipe Data Numerik

```python
data = {}
data["Jumlah Record"] = hasbi.shape[0]
data["Jumlah Attribute"] = hasbi.shape[1]
data["Nama Attribute"] = hasbi.columns.tolist()
data["Jumlah Attribute Dengan Tipe Angka"] = hasbi.select_dtypes(include=["int64", "float64"]).shape[1]

dictKeys = list(data.keys())

table = prtl.PrettyTable()
table.add_column(dictKeys[0], [data["Jumlah Record"]])
table.add_column(dictKeys[1], [data["Jumlah Attribute"]])
table.add_column(dictKeys[2], ['\n'.join(data['Nama Attribute'])])
table.add_column(dictKeys[3], [data["Jumlah Attribute Dengan Tipe Angka"]])
print(table)
```
