Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("salary.csv")
df.head()
```

Import library dan panggil csv yang kita gunakan

# Grafik Batang

```python
dfSex = df["sex"].value_counts()

plt.figure(figsize = (20, 15))
plt.bar(dfSex.index, dfSex, color = ["darkblue", "darkgreen"])
plt.xlabel("jenis kelamin", fontweight = "bold")
plt.ylabel("jumlah", fontweight = "bold")
plt.title("perbandingan jenis kelamin", fontsize = 20, fontweight = "bold")
plt.show() 
```

Dari grafik batang diatas kita bisa melihat bahwa lebih banyak laki laki dengan jumlah 20000an dan perempuan ada di angka sekitar 10000an

# Histogram

```python
plt.figure(figsize = (20, 15))
plt.hist(df["workclass"],color = "darkblue")
plt.xlabel("Kelas/Status Pekerjaan", fontsize = 15, fontweight = "bold")
plt.ylabel("Jumlah", fontsize = 15, fontweight = "bold")
plt.title("Tingkat Status Pekerjaan", fontsize = 20, fontweight = "bold")
plt.show() 
```

Bisa kita lihat di grafik histogram diatas bahwa sektor private(swasta) adalah jenis pekerjaan terbanyak dengan jumlah sekitar 20000an lebih dan sektor pekerjaan lainnya berada dibawah 5000an

# Pie Chart

```python
plt.figure(figsize = (10, 8))
dfSex = df["sex"].value_counts()
plt.pie(dfSex, labels = dfSex.index, autopct = "%1.1f%%", startangle = 90, shadow = True)
plt.title("Persentase Jenis Kelamin", fontsize = 20, fontweight = "bold")
plt.show() 
```

Bisa kita lihat untuk persentase jenis kelamin laki laki memiliki persentase sebanyak 66.9% sedangkan perempuan memiliki persentase 33.1% 

# Heatmap

```python
plt.figure(figsize = (8, 5))
sns.heatmap(df.corr(),
            annot = True)
plt.title("heatmap", fontsize = 20)
plt.show()
```

Korelasi yang mendominasi dan mendekati 1 adalah education-num
